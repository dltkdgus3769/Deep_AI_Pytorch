```python
from ultralytics import YOLO
import cv2

# YOLOv8 Î™®Îç∏ Î°úÎìú
# model = YOLO("yolov8n.pt")  # Nano Î™®Îç∏ ÏÇ¨Ïö©
# ÏÇ¨Ïö©Ïûê Ï†ïÏùò YOLOv8 Î™®Îç∏ Î°úÎìú
model = YOLO("./best-busanit501-aqua.pt")

# ÎπÑÎîîÏò§ ÌååÏùº Í≤ΩÎ°ú
video_path = "./Î∞îÎã§_Ìï¥ÏñëÎèôÎ¨ºÏÉòÌîåÏòÅÏÉÅ.mp4"

# Ï∂úÎ†• ÎπÑÎîîÏò§ ÏÑ§Ï†ï
output_path = "./output_Î∞îÎã§_Ìï¥ÏñëÎèôÎ¨º.mp4"

# ÎπÑÎîîÏò§ Ï∫°Ï≤ò Ï¥àÍ∏∞Ìôî
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Cannot open video file.")
    exit()

# ÎπÑÎîîÏò§ ÏÜçÏÑ± ÏùΩÍ∏∞
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*"mp4v")

# Ï∂úÎ†• ÎπÑÎîîÏò§ Ï¥àÍ∏∞Ìôî
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# ÌîÑÎ†àÏûÑ Îã®ÏúÑÎ°ú Í∞ùÏ≤¥ ÌÉêÏßÄ ÏàòÌñâ
frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:  # ÎπÑÎîîÏò§Ïùò ÎÅùÏóê ÎèÑÎã¨Ìïú Í≤ΩÏö∞
        break

    # YOLOv8ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Í∞ùÏ≤¥ ÌÉêÏßÄ
    results = model.predict(source=frame, save=False, conf=0.5)

    # ÌÉêÏßÄ Í≤∞Í≥º ÏãúÍ∞ÅÌôî
    annotated_frame = results[0].plot()

    # ÌÉêÏßÄ Í≤∞Í≥ºÎ•º Ï∂úÎ†• ÎπÑÎîîÏò§Ïóê Ï†ÄÏû•
    out.write(annotated_frame)

    # PyCharmÏóêÏÑú Ïã§ÏãúÍ∞Ñ ÌëúÏãú
    if frame_count % 30 == 0:
        cv2.imshow("YOLOv8 Detection", annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    frame_count += 1

# Î¶¨ÏÜåÏä§ Ìï¥Ï†ú
cap.release()
out.release()
print(f"Detection completed. Output video saved at {output_path}")
```
